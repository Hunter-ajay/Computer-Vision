```python
from ultralytics import YOLO
import cv2
import time
```

```python
model = YOLO("yolov8n.pt")
```

```python
import cv2
import time

cap = cv2.VideoCapture(0)  # open default webcam

if not cap.isOpened():
    print("Error: Could not open webcam.")
else:
    prev_time = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture frame")
            break

        # YOLO inference
        results = model(frame)
        annotated_frame = results[0].plot()

        # ----- FPS CALCULATION -----
        curr_time = time.time()
        fps = 1 / (curr_time - prev_time) if prev_time != 0 else 0
        prev_time = curr_time

        # Put FPS text on frame
        cv2.putText(
            annotated_frame,
            f"FPS: {fps:.1f}",
            (10, 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 255, 0),
            2,
            cv2.LINE_AA
        )

        # Show result
        cv2.imshow("YOLOv8 Detection - Press Q to exit", annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
